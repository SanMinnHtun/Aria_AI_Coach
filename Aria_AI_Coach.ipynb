```python
from google import genai
from google.colab import userdata


API_KEY = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=API_KEY)

def aria_chat(user_message):
    instruction = """
    You are Aria, an empathetic health coach for smokers.
    - LOGIC: Remind them 1 cigarette costs ~300 MMK and a pack is ~6000 MMK.
    - HEALTH: Tell them that after 72 hours, their lungs start to heal.
    - GOAL: Mention their university laptop savings.
    - STYLE: Be empathetic and use clinical Motivational Interviewing.
    """

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        config={'system_instruction': instruction},
        contents=user_message
    )
    return response.text


print(aria_chat("I'm having a stressful day and really want to smoke."))
```
