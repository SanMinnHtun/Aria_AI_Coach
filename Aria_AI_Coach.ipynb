```python
from google import genai
from google.colab import userdata

```

```python
API_KEY = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=API_KEY)
```

```python
def multi_agent_system(user_input):
    """
    Orchestrator: Manages specialized agents for health and savings.
    """

    master_instruction = """
    You are an AI Health System with three specialized agents:
    1. [Ko Chit Tee]: Financial expert. Calculates savings (300 MMK/cig) for things like laptops.
    2. [Dr.Yan Ku]: Biological expert. Focuses on physical recovery facts (e.g., 72-hour lung capacity).
    3. [ARI]: Empathy agent. Uses clinical MI techniques for stress and university exams.

    TASK: Identify the user's intent and respond ONLY as the relevant specialist.
    FORMAT: Start with '[Agent Name] AGENT RESPONDING'.
    """
    try:
        response = client.models.generate_content(
            model="gemini-3-flash-preview",
            config={'system_instruction': master_instruction},
            contents=user_input
        )
        return response.text
    except Exception as e:

        try:
            response = client.models.generate_content(
                model="gemini-2.5-flash",
                config={'system_instruction': master_instruction},
                contents=user_input
            )
            return response.text
        except:
            return f"System Error: {str(e)}"
```

```python
print(multi_agent_system("How much will I save if I quit for a month? I want a new laptop."))
```
